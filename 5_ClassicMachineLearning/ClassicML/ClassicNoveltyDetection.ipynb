##Parse Data

```
data_file = 'all_scaled0_1.csv'
```

```
data = pd.read_csv(data_file,index_col=0)
labels = data['label']
run = data['run']
chunkID = data['chunkID']
period = data['period']
data = data.drop(['label','chunkID','run','period'], axis = 1)
no_samples, no_features = data.shape
print(no_samples, no_features)
print(labels.value_counts())
data.head()
```

## Isolation Forest Outlier Detection

```
from sklearn.ensemble import IsolationForest

forest = IsolationForest(n_estimators = 100, contamination='auto',behaviour ='new')
pred = pd.Series(forest.fit_predict(data))
pred.value_counts()
```

**Evaluate by creating confusion matrix**

```
pred_s = pd.Series(np.where(pred == -1,1,0)) #Replace -1 -> 1, 1 -> 0 and turn to series
pred_s

evaluation = pd.Series(pred_s == labels)
confusion_matrix(labels, pred_s)
#  True pos.     False neg. 
#  False pos.    True neg.
```
