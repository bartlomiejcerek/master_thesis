```
#Read the list of sensitive features, strange formating so has to be done this way
with open('sensitive_variables_list.csv', 'r') as f:
    sensitive_txt = f.read().strip() #Dele whitespaces too
sensitive = sensitive_txt.split(',')

rank = pd.read_csv('feature_rank.csv',index_col=0)
logreg = rank['logreg'].to_list()
logreg_top100 = logreg[-100:]
```

```
in_both = [i for i in sensitive if i in logreg_top100]

print('Number of Sensitive Features: ', len(sensitive))
print('Number of Sensitive features in top 100 logreg: ', len(in_both))
```

```
#Take top 100 features and make union with sensitive ones
final_features = list(set(logreg_top100 + sensitive))
print('Final number of features: ', len(final_features))
```

```
# Save to file 
final_features = pd.Series(final_features)

final_features.to_csv("final_features_list.csv")
final_features.head()
```
